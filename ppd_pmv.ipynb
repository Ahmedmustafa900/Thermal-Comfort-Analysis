```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def calculate_ppd(pmv):
    """
    Calculate Predicted Percentage of Dissatisfied (PPD) from PMV.
    """
    return 100 - 95 * np.exp(-0.03353 * pmv**4 - 0.2179 * pmv**2)

# Step 1: Load the CSV file
file_name = r"C:\\Users\\AMUS\\python\\Yolov8_overall\\Yolo_updated_file.csv"  # Replace with your file name
data = pd.read_csv(file_name)

# Step 2: Define the columns
temperature_columns = ['HeadAll_Temperature']
vote_columns = ['head']

# Step 3: Iterate through temperature and vote columns
plt.figure(figsize=(10, 6))
for temp_col, vote_col in zip(temperature_columns, vote_columns):
    # Step 4: Calculate PMV and PPD for each pair
    data['PMV'] = data[vote_col]  # Assuming votes approximate PMV
    data['PPD'] = calculate_ppd(data['PMV'])
    
    # Step 5: Plot the data
    plt.plot(data[temp_col], data['PPD'], marker='o', label=f'{temp_col} vs PPD')

# Customize the plot
plt.xlabel('Temperature (C)')
plt.ylabel('Predicted Percentage of Dissatisfied (PPD) (%)')
plt.title('Temperature vs PPD for Different Body Parts')
plt.grid(True)
plt.legend()
plt.show()

# Optionally: Save the plot
# plt.savefig("temperature_vs_ppd_multiple.png")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def calculate_ppd(pmv):
    """
    Calculate Predicted Percentage of Dissatisfied (PPD) from PMV.
    """
    return 100 - 95 * np.exp(-0.03353 * pmv**4 - 0.2179 * pmv**2)

# Step 1: Load the CSV file
file_name = r"C:\\Users\\AMUS\\python\\Yolov8_overall\\Yolo_updated_file.csv"  # Replace with your file name
data = pd.read_csv(file_name)

# Step 2: Define the columns
temperature_columns = ['HeadAll_Temperature']
vote_columns = ['head']

# Step 3: Calculate PMV and PPD for each pair
data['PMV'] = data[vote_columns[0]]  # Assuming votes approximate PMV
data['PPD'] = calculate_ppd(data['PMV'])

# Step 4: Get unique temperature values for boxplot grouping
unique_temperatures = data[temperature_columns[0]].unique()

# Step 5: Create a box plot for PPD distribution across different temperature values
plt.figure(figsize=(12, 8))

# Collect the PPD values for each unique temperature and plot the box plot
ppd_values_by_temp = [data[data[temperature_columns[0]] == temp]['PPD'] for temp in unique_temperatures]

# Box plot
plt.boxplot(ppd_values_by_temp, labels=[f'{temp} C' for temp in unique_temperatures])

# Customize the plot
plt.xlabel('Temperature (C)', fontsize=14)
plt.ylabel('Predicted Percentage of Dissatisfied (PPD) (%)', fontsize=14)
plt.title('Temperature vs PPD for Head - Box Plot', fontsize=16, fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0, 100)  # PPD ranges from 0% to 100%

# Display the plot
plt.tight_layout()
plt.show()

# Optionally: Save the plot as a file
# plt.savefig("temperature_vs_ppd_head_boxplot.png", dpi=300)
```
